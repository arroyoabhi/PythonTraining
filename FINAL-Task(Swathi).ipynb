## TASK-4
### Donut charts

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randint(10, 100, 6)
plt.pie(data, labels = ['A', 'B', 'C', 'D', 'E', 'F'])

circle = plt.Circle( (0,0), 0.6, color='yellow')


plt.title('DONUT CHART')

p=plt.gcf()
p.gca().add_artist(circle)
plt.show()
```

### Donut chart-Subject Details

```python
import matplotlib.pyplot as plt

labels = 'Maths', 'English', 'Science', 'Hindi'
sizes = [15, 30, 45, 10]
colors = ['yellow', 'orange', 'blue', 'green']
explode=(0, 0,0,0)

plt.title("No. of Students per subject",fontsize=20)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%2.2f%%')
        

centre_circle = plt.Circle((0,0),0.5,color='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.show()  
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("data.csv")

explode = [0, 0.1, 0]
colors = ["green", "blue","red"]
labels = [60,45,30]

plt.title("Donut chart of Duration column",fontsize=25)

plt.pie(df['Duration'].value_counts(),labels=labels,explode = explode,colors = colors, shadow = True);
plt.legend(title = "Duration:")

centre= plt.Circle((0,0),0.5,color='black',fc='white', linewidth=3)
fig = plt.gcf()
fig.gca().add_artist(centre)

plt.show()
```
