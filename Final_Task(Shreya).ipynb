# Task-1
- add the above two arrays
- multiply each element of the array with 2 after addition

```python
age1 = [2,3,4,6,5]
age2= [4,6,7,8,9]
print(f'List of elements in age1: {age1}\n')
print(f'List of elements in age2: {age2}\n')

addAge=np.add(age1, age2) 
print(f'After adding above 2 arrays: {addAge}\n')

mul = np.array(addAge)
print (f'After multiplying each element with 2: {mul * 2}')
```

```python
age1 = [2,3,4,6,5]
age2= [4,6,7,8,9]
  
print ("List of elements in age1: ",age1)
print ("\nList of elements in age2: ",age2)
  

age = []
for i in range(0, len(age1)):
    age.append(age1[i] + age2[i])
print ("\nAfter adding above 2 arrays : ",age)

mul = [x*2 for x in age]
print("\nAfter multiplying each element with 2 :",mul)
```

# Task-3
### Handling dates
- Find how much time it takes to place an order and ship the product
- How many times has it taken more than 5 days from placing an order to shipping

```python
data=pd.read_csv('Store.csv',header=0,encoding="latin")
data.head()
```

### Converting all the dates to date format

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data
```

### Adding a seperate column 'Shipping Days' containing the number of days from placing an order till shipping the product

```python
data['ShippingDays'] = (data['Ship Date'] - data['Order Date'])
data.head()
```

### Converting the Shipping days from timedelta to int data type.

```python
data['ShippingDays'] = (data['ShippingDays']).dt.days
data.head()
```

```python
#data['Shipping Days'] = (data['Ship Date'] - data['Order Date']).dt.days
#data.head()
```

### Number of times it taken more than 5 days from placing an order to shipping

```python
c=data['ShippingDays'] > 5
c.value_counts()
```

```python
data[data.ShippingDays > 5].count() 
```

```python
data[data['ShippingDays'] > 5]
```
